```python
import rasterio
import numpy as np
import os
from rasterio.merge import merge

def get_metadata(mtl_path):
    params ={}
    with open(mtl_path,'r') as mtl: # doc du lieu trong mtl luu vao trong params
        for line in mtl:
            if '=' in line:
                key, value = line.split('=')
                key =  key.strip()
                value = value.strip().replace('"','')
                try:
                    value = float(value)
                except ValueError:
                    pass 
                params[key] = value
        return params



# doi sang nhiet do don vi C
def cal_temperature(b10_path,mtl_path):
    params = get_metadata(mtl_path)
    scale_factor = params['TEMPERATURE_MULT_BAND_ST_B10']
    offset = params['TEMPERATURE_ADD_BAND_ST_B10']
    # print(scale_factor)
    # print(offset)
    with rasterio.open(b10_path) as src:
        
        dataset = src.read(1).astype(np.float32)

        # xu ly pixel bi may che phu
        qa_path = b10_path.replace('B10','QA').replace('image','qa_file')

        mask = np.zeros_like(dataset, dtype=bool)

        # with rasterio.open(qa_path) as qa_src:
        #     qa = qa_src.read(1).astype(np.uint16)
        #     qa_mask = (qa & int('00000011', 2)) != 0
        #     mask |= qa_mask
        #     dataset[mask] = np.nan

        profile  = src.profile
        # print(profile)
        dataset[dataset == src.nodata  ] = np.nan
        ds = dataset*scale_factor + offset - 273.15
        
        ds[ds<25] = np.nan
        profile.update(
            dtype = 'float32',
            nodata = np.nan,
        )
    out_path = b10_path.replace('B10','B10_TEMP')
    if not os.path.exists(out_path):
        with rasterio.open(out_path,'w',**profile) as dst:
            dst.write(ds,1)
        
    return ds
    

    
def merge_raster(year,out_path):
   # out_path = 'out_image/LST_' +year+ '.TIF'
    src_path1 = out_path.replace(('out_image/LST_'+year),('image/LC08_L2SP_127045_' + year + '_02_T1_ST_B10_TEMP'))
    src_path2 = out_path.replace(('out_image/LST_'+year),('image/LC08_L2SP_127046_' + year + '_02_T1_ST_B10_TEMP'))

    srcs=[]
    src_paths = [src_path1,src_path2]
    for path in src_paths:
        if not os.path.exists(path):
            cal_temperature(path.replace('B10_TEMP','B10'),path.replace('image/','metadata/').replace('ST_B10_TEMP','MTL').replace('.TIF','.txt'))
            
      
   

    if not os.path.exists(out_path): 
        with rasterio.open(src_path1) as src1, rasterio.open(src_path2) as src2:
            profile = src1.profile.copy()
            dest_ds,out_trans = merge([src1,src2])
            
            dest_ds = dest_ds.copy()
            print(dest_ds.shape)

            profile.update({
                "count" : 1,
                "dtype" : 'float32',
                "nodata" : np.nan,
                "height" : dest_ds.shape[1],
                "width" : dest_ds.shape[2],
                "transform" : out_trans,
            })

            with rasterio.open(out_path,'w',**profile) as dst:
                dst.write(dest_ds)

    

    os.remove(src_path1)
    os.remove(src_path2)
    
```

```python
from rasterstats import zonal_stats
import fiona
import rasterio.mask
```

```python

year = ['2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
#year =['2019','2020','2021','2023']
average_hn =[]
average_ngoaithanh = []
average_noithanh = []
averages =[]
averages.append(average_hn)
averages.append(average_noithanh)
averages.append(average_ngoaithanh)

for y in year:
    if not os.path.exists('out_image/LST_' + y + '.TIF'):
        merge_raster(y,'out_image/LST_' + y + '.TIF')
    if not os.path.exists('out_image/LST_' + y + '(1).TIF'):
        merge_raster(y,'out_image/LST_' + y + '(1).TIF')

    shapefiles = []
    
    
    with fiona.open('shapefile/HN.shp', "r") as shapefile_HN,\
    fiona.open('shapefile/NOIDO_HN.shp', "r") as shapefile_ND\
    ,fiona.open('shapefile/NGOAI_THANH_HN.shp', "r") as shapefile_NT:

        shape_HN = [feature["geometry"] for feature in shapefile_HN]
        shape_ND = [feature["geometry"] for feature in shapefile_ND]
        shape_NT = [feature["geometry"] for feature in shapefile_NT]
        shapefiles.append(shape_HN)
        shapefiles.append(shape_ND)
        shapefiles.append(shape_NT)
    path1 = 'out_image/LST_' + y + '.TIF'
    path2 = 'out_image/LST_' + y + '(1).TIF'

    for i in range(3):
        with rasterio.open(path1) as src1:
            out_image1, out_transform1 = rasterio.mask.mask(src1, shapefiles[i], crop=True)
            out_meta = src1.meta
        with rasterio.open(path2) as src2:
            out_image2, out_transform2 = rasterio.mask.mask(src2, shapefiles[i], crop=True)
            out_meta = src1.meta
  
        out_meta.update({"driver": "GTiff",
                    "height": out_image1.shape[1],
                    "width": out_image1.shape[2],
                    "transform": out_transform1})
        
        stacked = np.stack([out_image1, out_image2], axis=0)
        aver2image = np.nanmean(stacked, axis=0)

        
        with rasterio.open('out_image/LST_'+y+str(i)+'clipped.TIF', "w", **out_meta) as dest:
            dest.write(aver2image)

        averages[i].append(np.nanmean(aver2image))

#     stats_hn = zonal_stats('shapefile/HN.shp','out_image/LST_' + y + '.TIF',stats='mean')
#     stats_ngoaithanh = zonal_stats('shapefile/NGOAI_THANH_HN.shp','out_image/LST_' + y + '.TIF',stats='mean')
#     stats_noithanh = zonal_stats('shapefile/NOIDO_HN.shp','out_image/LST_' + y + '.TIF',stats='mean')
#     average_hn.append(stats_hn[0]['mean'])
#     average_ngoaithanh.append(stats_ngoaithanh[0]['mean'])
#     average_noithanh.append(stats_noithanh[0]['mean'])

# print(average_hn)
# print(average_ngoaithanh)
# print(average_noithanh)


```

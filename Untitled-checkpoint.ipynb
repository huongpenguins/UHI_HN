```python
import rasterio
import numpy as np
import os
from rasterio.merge import merge
import rasterio
import numpy as np
import os
from rasterio.merge import merge
from rasterstats import zonal_stats
import fiona
import rasterio.mask

# tim anh theo nam va thang
# tra ve danh sach 2 anh ung voi 2 ngay trong nam va thang 
def find_b10_file(path_row, year, month):
    folder = 'image'
    paths=[]
    for f in os.listdir(folder):
        if  f.endswith('ST_B10.TIF') and path_row in f:
            date_str = f.split('_')[3]  
            if date_str.startswith(year + month):
                file = os.path.join(folder, f)
                paths.append(file)
    return paths  
```

```python
# lay thong tin tu file metadata
# tra ve scale va offset ung voi anh b10
def get_metadata(mtl_path):
    params ={}
    with open(mtl_path,'r') as mtl: # doc du lieu trong mtl luu vao trong params
        for line in mtl:
            if '=' in line:
                key, value = line.split('=')
                key =  key.strip()
                value = value.strip().replace('"','')
                try:
                    value = float(value)
                except ValueError:
                    pass 
                params[key] = value
        return params



# doi sang nhiet do don vi C tu thong tin metadata

def cal_temperature(b10_path,mtl_path):

    params = get_metadata(mtl_path)
    scale_factor = params['TEMPERATURE_MULT_BAND_ST_B10']
    offset = params['TEMPERATURE_ADD_BAND_ST_B10']

 
    with rasterio.open(b10_path) as src:
        # xu ly pixel bi may che phu

        qa_path = b10_path.replace('ST_B10','QA_PIXEL').replace('image','qa_file')

        if not os.path.exists(qa_path):
            if('T1' in qa_path):
                qa_path=qa_path.replace('T1','T2')
            else:
                qa_path=qa_path.replace('T2','T1')

        with rasterio.open(qa_path) as qa_src:

        # cat anh ban dau theo khu vuc ha noi
            with fiona.open('shapefile/HN.shp', "r") as shapefile_HN:
                shape_HN = [feature["geometry"] for feature in shapefile_HN]
                out_image, out_transform = rasterio.mask.mask(src, shape_HN, crop=True)
                out_image = np.where(out_image == src.nodata, -9999, out_image)
                data = out_image[0].astype(np.float32)
                qa_clipped, qa_out_transform = rasterio.mask.mask(qa_src, shape_HN, crop=True)
                qa = qa_clipped[0].astype(np.uint16)
             

        qa_mask = ((qa & (1 << 0)) != 0) | \
                    ((qa & (1 << 2)) != 0) | \
                    ((qa & (1 << 3)) != 0) | \
                    ((qa & (1 << 4)) != 0) | \
                    ((qa & (1 << 7)) != 0)
        
        profile = src.profile   
        mask = np.zeros_like(data, dtype=bool)
        mask |= qa_mask
        
        data[mask] = -9999
        ds= data
        ds = data*scale_factor + offset - 273.15
        ds[ds<15] = -9999
        ds[ds>70] = -9999

        profile  = src.profile.copy()
        profile.update({
            "height" : out_image.shape[1],
            "width" : out_image.shape[2],
            "transform" : out_transform,
            "dtype" : 'float32',
            "nodata" : -9999
        })

    out_path = b10_path.replace('B10','B10_TEMP')
    if not os.path.exists(out_path):
        with rasterio.open(out_path,'w',**profile) as dst:
            dst.write(ds,1)
        
    return ds
    
# ghep 2 anh de lay duoc toan bo khu vuc ha noi
def merge_raster(out_path,b10_temp_path):
    
        src_path1 = b10_temp_path
# doi tu T1 sang T2 hoac nguoc lai neu can
        src_path2 = b10_temp_path.replace('127045','127046')
        if not os.path.exists(src_path2.replace('B10_TEMP','B10')):
            if('T1' in src_path2):
                src_path2 = src_path2.replace('T1','T2')
            else:
                src_path2 = src_path2.replace('T2','T1')
        
        src_paths = [src_path1,src_path2]

        for path in src_paths:
            if not os.path.exists(path):
                mtl = path.replace('image','metadata').replace('ST_B10_TEMP','MTL').replace('.TIF','.txt')
                if not os.path.exists(mtl):
                    if('T1' in mtl):
                        mtl=mtl.replace('T1','T2')
                    else:
                        mtl=mtl.replace('T2','T1')
                cal_temperature(path.replace('B10_TEMP','B10'),mtl)
                
        
    

        if not os.path.exists(out_path): 
            with rasterio.open(src_path1) as src1, rasterio.open(src_path2) as src2:
                profile = src1.profile.copy()
                dest_ds,out_trans = merge([src1,src2])
                
                dest_ds = dest_ds.copy()
                print(dest_ds.shape)

                profile.update({
                    "count" : 1,
                    "dtype" : 'float32',
                    "nodata" : -9999,
                    "height" : dest_ds.shape[1],
                    "width" : dest_ds.shape[2],
                    "transform" : out_trans,
                })

                with rasterio.open(out_path,'w',**profile) as dst:
                    dst.write(dest_ds)

        

        os.remove(src_path1)
        os.remove(src_path2)
    
```

```python

months = ['07']
year = ['2014','2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']
# year =['2020']
average_hn =[]
average_ngoaithanh = []
average_noithanh = []
averages =[]
averages.append(average_hn)
averages.append(average_noithanh)
averages.append(average_ngoaithanh)
for month in months:
    for y in year:
        paths = []
        b10_paths = find_b10_file('127045',y,month)
        for p in b10_paths:
            filename = os.path.basename(p)       
            day = filename.split('_')[3][6:]
            paths.append('out_image/LST_' + y + month + day + '.TIF')
        

        path1 = paths[0]
        path2 = paths[1]
        if not os.path.exists(path1):
            merge_raster(path1,b10_paths[0].replace('B10','B10_TEMP'))
        if not os.path.exists(path2):
            merge_raster(path2,b10_paths[1].replace('B10','B10_TEMP'))
        shapefiles = []
        
        
        with fiona.open('shapefile/HN.shp', "r") as shapefile_HN,\
        fiona.open('shapefile/NOIDO_HN.shp', "r") as shapefile_ND\
        ,fiona.open('shapefile/NGOAI_THANH_HN.shp', "r") as shapefile_NT:

            shape_HN = [feature["geometry"] for feature in shapefile_HN]
            shape_ND = [feature["geometry"] for feature in shapefile_ND]
            shape_NT = [feature["geometry"] for feature in shapefile_NT]
            shapefiles.append(shape_HN)
            shapefiles.append(shape_ND)
            shapefiles.append(shape_NT)
        

        for i in range(3):
            with rasterio.open(path1) as src1:
                out_image1, out_transform1 = rasterio.mask.mask(src1, shapefiles[i], crop=True)
                out_image1 = np.where(out_image1 == -9999, np.nan, out_image1)
                out_meta = src1.meta.copy()
            with rasterio.open(path2) as src2:
                out_image2, out_transform2 = rasterio.mask.mask(src2, shapefiles[i], crop=True)
                out_image2 = np.where(out_image2 == -9999, np.nan, out_image2)
                
        
            out_meta.update({"driver": "GTiff",
                        "height": out_image1.shape[1],
                        "width": out_image1.shape[2],
                        "transform": out_transform1,
                        "count": 1
                        })
            
            
            stacked = np.stack([out_image1, out_image2], axis=0)
            aver2image = np.nanmean(stacked, axis=0)

            averages[i].append(np.nanmean(aver2image))
            
            aver2image = np.where(np.isnan(aver2image), -9999, aver2image)

            with rasterio.open('out_image/LST_'+ y + month + day + str(i) + 'clipped.TIF', "w", **out_meta) as dest:
                dest.write(aver2image)

            

    #     stats_hn = zonal_stats('shapefile/HN.shp','out_image/LST_' + y + '.TIF',stats='mean')
    #     stats_ngoaithanh = zonal_stats('shapefile/NGOAI_THANH_HN.shp','out_image/LST_' + y + '.TIF',stats='mean')
    #     stats_noithanh = zonal_stats('shapefile/NOIDO_HN.shp','out_image/LST_' + y + '.TIF',stats='mean')
    #     average_hn.append(stats_hn[0]['mean'])
    #     average_ngoaithanh.append(stats_ngoaithanh[0]['mean'])
    #     average_noithanh.append(stats_noithanh[0]['mean'])

    # print(average_hn)
    # print(average_ngoaithanh)
    # print(average_noithanh)


```
